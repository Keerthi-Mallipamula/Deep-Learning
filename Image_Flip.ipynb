```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
india = cv2.imread(r"C:\Users\user\Desktop\india-map.jpg")
```

```python
india
```

```python
cv2.imshow("India Map", india)
cv2.waitKey(-1)
```

```python
india.shape
```

```python
bharat = cv2.resize(india,(400, 550))
```

```python
bharat
```

```python
cv2.imshow("India Map", india)
cv2.imshow("Bharat Map", bharat)
cv2.waitKey(-1)
```

```python
cv2.imshow("Bharat Map", bharat)
cv2.waitKey(-1)
```

```python
bharat_0 = cv2.flip(bharat, 0)
```

```python
cv2.imshow("Bharat Flip Zero", bharat_0)
cv2.waitKey(-1)
```

```python
cv2.imshow("Bharat Map", bharat)
cv2.imshow("Bharat Flip Zero", bharat_0)
cv2.waitKey(-1)
```

```python
bharat_1 = cv2.flip(bharat, 1)
```

```python
cv2.imshow("Bharat Flip One", bharat_1)
cv2.waitKey(-1)
```

```python
cv2.imshow("Bharat Map", bharat)
cv2.imshow("Bharat Flip Zero", bharat_0)
cv2.imshow("Bharat Flip One", bharat_1)
cv2.waitKey(-1)
```

```python
bharat2 = cv2.flip(bharat, -1)
```

```python
cv2.imshow("Bharat_Flip_2", bharat2)
cv2.waitKey(-1)
```

```python
cv2.imshow("Bharat Map", bharat)
cv2.imshow("Bharat Flip Zero", bharat_0)
cv2.imshow("Bharat Flip One", bharat_1)
cv2.imshow("Bharat_Flip_2", bharat2)
cv2.waitKey(-1)
```
